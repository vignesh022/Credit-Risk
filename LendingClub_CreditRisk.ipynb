### Target variable

The choice of target variable is key to our analysis as we are looking to predict, if based on a given set of features, a borrower is a good for the loan amount.

The only variable that fits the bill is "loan_status" as it describes the activity of the borrower w.r.t the loan i.e. it directly describes if a loan was paid off on time, had delayed payments, or was defaulted on the borrower.

```python
loans_data['loan_status'].value_counts(normalize=True)*100
```

~78% of the loans offered via Lending Club have been paid off in full. Only ~0.01% of the loans are defaulted upon.

The meaning behind each status message was learnt through their [webpage](https://help.lendingclub.com/hc/en-us/articles/215488038-What-do-the-different-Note-statuses-mean-). 

| Loan Status |	Meaning |
| ----------- | ------- |
| Fully Paid	|	Loan has been fully paid off. |
| Charged Off	|	Loan for which there is no longer a reasonable expectation of further payments. |
| Does not meet the credit policy. Status:Fully Paid	|	The loan was paid off, application wouldn't be approved on to the marketplace in the present. |
| Does not meet the credit policy. Status:Charged Off | The loan was charged off, application wouldn't be approved on to the marketplace in the present. |
| In Grace Period	|	The loan is past due but still in the grace period of 15 days. |
| Late (16-30 days)	|	Loan hasn't been paid in 16 to 30 days (late on the current payment). |
| Late (31-120 days)	|	Loan hasn't been paid in 31 to 120 days (late on the current payment). |
| Current	|	Loan is up to date on current payments. |
| Default	|	Loan is defaulted on and no payment has been made for more than 121 days. 



Since the risk prediction for a potential new loan can only be a "Yes" or "No", the problem is essentially a binary classification. The status messages that best reflect the outcomes desired are "Fully Paid" and "Charged Off", both of these messages imply that there is no expectation for any future payments no matter how marginal the chances.

We adopt a binary notation moving forward, "Fully Paid" equals 1 and "Charged Off" is equal to 0.

```python
loans_data = loans_data[(loans_data['loan_status'] == "Fully Paid") | (loans_data['loan_status'] == "Charged Off")]

status_replace = {
    "loan_status" : {
        "Fully Paid": 1,
        "Charged Off": 0,
    }
}

loans_data = loans_data.replace(status_replace)
```

We can also drop columns with a lack of unique values. These columns don't change across the dataset and hence would not add any valuable information to our modeling.

```python
original_columns = loans_data.columns
drop_columns = []
for col in original_columns:
    col_series = loans_data[col].dropna().unique()
    if len(col_series) == 1:
        drop_columns.append(col)

print(drop_columns)
clean_loans_data = loans_data.drop(drop_columns, axis=1)
```

```python
clean_loans_data.isnull().sum()
```

Looking at the null values in the columns we have retained moving forward, 2 columns stand out from the rest.

"emp_length" which is the number of years the borrower has been employed at a given job, it is regarded as a key identifier of the financial stability of the borrower and must be retained.

```python
print(clean_loans_data.pub_rec_bankruptcies.value_counts(normalize=True, dropna=False))
```

Close to ~94% of the column is filled by 0, making this column another candidate to be dropped completely as it adds no value to our predictive analysis.

```python
clean_loans_data = clean_loans_data.drop("pub_rec_bankruptcies", axis=1)
clean_loans_data = clean_loans_data.dropna(axis=0)
```

The "title" column is a string dtype column with the reasoning behind the borrowers application for a loan. There are more than 18000+ reasons provided by the borrowers for their loan requests.

Debt consolidation appears  to be a popular reason for requesting loans.

The "emp_length" column presents data as a string but can be converted to a numerical column.

```python
mapping_dict = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
        "n/a": 0
    }
}
clean_loans_data = clean_loans_data.drop(["last_credit_pull_d", "earliest_cr_line", "addr_state", "title"], axis=1)
clean_loans_data["int_rate"] = clean_loans_data["int_rate"].str.rstrip("%").astype("float")
clean_loans_data["revol_util"] = clean_loans_data["revol_util"].str.rstrip("%").astype("float")
clean_loans_data = clean_loans_data.replace(mapping_dict)
```

### Dummy coding categorical variables

```python
cat_columns = ["home_ownership", "verification_status", "purpose", "term"]

dummy_df = pd.get_dummies(clean_loans_data[cat_columns])
clean_loans_data = pd.concat([clean_loans_data, dummy_df], axis=1)
clean_loans_data = clean_loans_data.drop(cat_columns, axis=1)
```

## Modeling, Analysis, and Predicting

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

logit_reg = LogisticRegression(solver='liblinear')
cols = clean_loans_data.columns
train_cols = cols.drop("loan_status")
features = clean_loans_data[train_cols]
target = clean_loans_data["loan_status"]
logit_reg.fit(features, target)
predictions = logit_reg.predict(features)
```

```python
from sklearn.model_selection import cross_val_predict
logit_reg = LogisticRegression(solver='liblinear')
predictions = cross_val_predict(logit_reg, features, target, cv=5)
predictions = pd.Series(predictions)

# False positives.
fp_filter = (predictions == 1) & (clean_loans_data["loan_status"] == 0)
fp = len(predictions[fp_filter])

# True positives.
tp_filter = (predictions == 1) & (clean_loans_data["loan_status"] == 1)
tp = len(predictions[tp_filter])

# False negatives.
fn_filter = (predictions == 0) & (clean_loans_data["loan_status"] == 1)
fn = len(predictions[fn_filter])

# True negatives
tn_filter = (predictions == 0) & (clean_loans_data["loan_status"] == 0)
tn = len(predictions[tn_filter])
# Rates
tpr = tp  / (tp + fn)
fpr = fp  / (fp + tn)
print(tpr)
print(fpr)
```

An issue that must be dealt with in our analysis is the class imbalance between the "Fully Paid" and "Charged Off" labels, their ratio is almost 6:1 i.e. for every one loan that is "Charged Off" there are 6 loans fully paid.

We can balance by weighting the dataset appropriately.

```python
logit_reg = LogisticRegression(class_weight="balanced",solver='liblinear')
predictions = cross_val_predict(logit_reg, features, target, cv=5)
predictions = pd.Series(predictions)

# False positives.
fp_filter = (predictions == 1) & (clean_loans_data["loan_status"] == 0)
fp = len(predictions[fp_filter])

# True positives.
tp_filter = (predictions == 1) & (clean_loans_data["loan_status"] == 1)
tp = len(predictions[tp_filter])

# False negatives.
fn_filter = (predictions == 0) & (clean_loans_data["loan_status"] == 1)
fn = len(predictions[fn_filter])

# True negatives
tn_filter = (predictions == 0) & (clean_loans_data["loan_status"] == 0)
tn = len(predictions[tn_filter])

# Rates
tpr = tp / (tp + fn)
fpr = fp / (fp + tn)

print(tpr)
print(fpr)
```
